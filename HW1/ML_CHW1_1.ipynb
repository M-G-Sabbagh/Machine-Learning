```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def coefficients(X,Y):
    
    A = X @ np.transpose(X)
    B = X @ Y
    if np.shape(A) == ():
        W = A**(-1) * Y
    else:
        W = np.linalg.inv(A) @ B
    
    return W

def estimation(W,X):
    return np.transpose(W) @ X

def MSE(W,X,Y):
    
    num_columns = np.shape(X)[1]
    error = 0
    est = estimation(W,X)
    
    for i in range(num_columns):
        error = error + (est[i] - Y[i])**2
    error = error / num_columns 
    return error
    
```

```python
dataset = pd.read_csv("Q1_data.csv")
```

```python
npdataset = dataset.to_numpy()
```

```python
tdata = npdataset[0:402, 1:]
vdata = npdataset[402:, 1:]

#Training data
Yt = tdata[:, 7]
Xt = tdata[:, 0:7]
Xt = np.transpose(Xt)
one = np.ones(shape = 402)
Xt = np.vstack((one, Xt))
W = coefficients(Xt,Yt)
tdataerror = MSE(W,Xt,Yt)
est = estimation(W,Xt)
print("W = ", W)
print("Empirical Risk = ", tdataerror)

#Validation data
Yv = vdata[:, 7]
Xv = vdata[:, 0:7]
Xv = np.transpose(Xv)
one = np.ones(shape = 100)
Xv = np.vstack((one, Xv))
vdataerror = MSE(W,Xv,Yv)

print("True Risk = ",vdataerror)
```

```python
# Feature 1:
x1 = Xt[0:2, :]
w1 = coefficients(x1,Yt)
y1 = estimation(w1,x1)

# Feature 2:
one = np.ones(shape = 402)
x2 = np.vstack((one, Xt[2, :]))
w2 = coefficients(x2,Yt)
y2 = estimation(w2,x2)

# Feature 3:
x3 = np.vstack((one, Xt[3, :]))
w3 = coefficients(x3,Yt)
y3 = estimation(w3,x3)

# Feature 4:
x4 = np.vstack((one, Xt[4, :]))
w4 = coefficients(x4,Yt)
y4 = estimation(w4,x4)

# Feature 5:
x5 = np.vstack((one, Xt[5, :]))
w5 = coefficients(x5,Yt)
y5 = estimation(w5,x5)

# Feature 6:
x6 = np.vstack((one, Xt[6, :]))
w6 = coefficients(x6,Yt)
y6 = estimation(w6,x6)

# Feature 7:
x7 = np.vstack((one, Xt[7, :]))
w7 = coefficients(x7,Yt)
y7 = estimation(w7,x7)

plt.plot(np.arange(402), Yt, 'o', color='black')
plt.plot(np.arange(402), y1, 'o', color='blue')
plt.plot(np.arange(402), y2, 'o', color='red')
plt.plot(np.arange(402), y3, 'o', color='green')
plt.plot(np.arange(402), y4, 'o', color='yellow')
plt.plot(np.arange(402), y5, 'o', color='orange')
plt.plot(np.arange(402), y6, 'o', color='purple')
plt.plot(np.arange(402), y7, 'o', color='pink')
```
